```python
def assign_segment_from_list(df: pd.DataFrame, segment_list: list) -> pd.DataFrame:
    """
    Überprüft, ob ein Segmentname aus der gegebenen Liste in der Spalte
    'Modell_Segment_Raw' des DataFrames vorhanden ist und schreibt den
    gefundenen Segmentnamen in eine neue Spalte 'Gefundenes_Segment'.

    Priorisiert längere Segmentnamen, um spezifischere Treffer zu bevorzugen.
    Die Suche ist nicht-case-sensitiv.

    Args:
        df (pd.DataFrame): Der DataFrame mit der Spalte 'Modell_Segment_Raw'.
        segment_list (list): Eine Liste von eindeutigen Segmentnamen.

    Returns:
        pd.DataFrame: Der DataFrame mit der neuen Spalte 'Gefundenes_Segment'.
    """
    # Erstelle eine Kopie des DataFrames, um SettingWithCopyWarning zu vermeiden
    df_copy = df.copy()

    # Initialisiere die neue Spalte mit einem Standardwert (z.B. None oder leeren String)
    df_copy['Gefundenes_Segment'] = None

    # Sortiere die Segmentliste nach Länge absteigend, um spezifischere Matches zu bevorzugen.
    # Dies ist nützlich, wenn ein Segment ein Substring eines anderen ist (z.B. "SUV" und "Kompakt-SUV").
    sorted_segment_list = sorted(segment_list, key=len, reverse=True)

    # Iteriere über jede Zeile des DataFrames
    for index, row in df_copy.iterrows():
        # Stelle sicher, dass der Inhalt ein String ist und konvertiere ihn zu Kleinbuchstaben für die Suche
        raw_string = str(row['Modell_Segment_Raw']).lower()

        found_segment = None
        for segment in sorted_segment_list:
            # Überprüfe, ob der Segmentname (kleingeschrieben) im Rohstring enthalten ist
            if segment.lower() in raw_string:
                found_segment = segment # Speichere den Original-Segmentnamen aus der Liste
                break # Den ersten (und damit längsten/spezifischsten) Treffer nehmen

        df_copy.at[index, 'Gefundenes_Segment'] = found_segment

    return df_copy
```
